## Description of the cohort/ behaviour:
### Mice:
8 mice, males, wild type, unmodified.
### Goal of the training:
To establish the learning parameters and performance of wild type mice in the full behaviour.
### Behaviour:
Mice were trained with the end goal of achieving one perticular task:
Mice self initiate each trial by standing on a small platform/ puck in the centre of the red hex arena.
This platform is a set of scales which detects the weight of the mouse to automatically determine the presence of the mouse. 
At this point, a cue is given, and mice must decide which port to go to to receive some milk reward.
The cues can either be visual or audio based.
In visual trials, a green LED flashes on one of the 6 walls, indicating to the mouse that that port will give reward.
In audio trials, an overhead speaker beeps, indicating that mice must go to port 1 to receive their reward.
The duration of each LED and audio cue is identical.
If mice go to the correct port, they receive a drop of milk, and can immediately go back to the platform to self initiate a new trial.
If mice go to the wrong port, an error signal is given and mice receive nothing. The error signal consists of a 10 second spotlight brightness increase, and mice are not able to initiate a new trial within this 10 seconds, acting as a timeout.
If mice do not go to a port, the trial times out after 10 seconds, after which a new cue is given if the mouse activates the scales.
Mice then continue to perform trials for the duration of the session, usually about 1hr long. In this time mice are able to perform up to 400 trials.
### Training:
Mice have their behaviour shaped to achieve this end task. This is done by moving through different phases of the behaviour over the course of a week, starting out with simple versions of the task and progressively making it harder. In this cohort, progression onto the next phase was judged manually. Phase numbers/ names are a bit messy.
- **Phase 2:**
*Aim*: To teach the mouse to go to the platform to receive reward. 
*Task*: When mice put their full weight on the platform, a drop of milk is immediately given at port 1, which the mouse can then go and find. Once the mouse has taken the reward (activated the sensor at port 1), they are able to get another drop by going back to the platform. 
*Performance evaluation*: Mice are moved on to the next phase when they are consistently going back and forth between the platform and the reward port, demonstrating that they understand that visiting the platform causes reward release at a specific port.
*Early behaviour*: Mice will explore the arena, investigating ports to see if there's milk, as usually they are able to find residue in the arena. Mice are encouraged onto the platform by a drop of milk being put in the centre by the experimenter, so after the mice find this they are more likely to revisit the platform and cause a port 1 activation. Once they are visiting the platform, it takes a bit for them to realise the reward is always at port 1, meaning they will oftern visit other ports first in a semi-random manner.
- **Phase 3:**
*Aim*: To teach mice to only visit port 1, and also to introduce the LED and that the reward is only given when the mouse nose-pokes.
*Task*: Mice initiate a trial by standing on the platform, at which point the LED at port 1 comes on. If the mice visit port 1 then they receive the reward, however if they visit another port first then an error signal is given and they receive no reward that trial. The LED turns off when the mouse touches a port.
*Performance evaluation*: Mice move on to the next phase when they are consistently getting the reward and not getting error signals at a high rate.
- **Phase 4:**
*Aim*: To teach mice to wait for a short time on the platform before the cue is given. Pause time = 1s.
*Task*: This is the same as phase 3 except that mice must pause for a short time on the platform before the cue is given. This is to encourage waiting for the cue to be given so that mice don't impulsively walk off the platform before seeing where the cue is. 
*Performance evaluation*: Mice are moved on when they can consistently get the reward and are not getting errors. They should also be showing a clear pause on the platform before moving off. 
- **Phase 'test':**
*Aim*: To examine the strategy that the mice are using to get to port 1. 
*Task*: 5 trials are given with the LED at port 1 and then 10 trials are given with the LED at port 3. IF the mice go to either port 1 or 3 they can receive reward.
*Performance evaluation*: The aim is to evaluate if the mice are using the LED or the external landmarks in the arena to navigate to port 1. If mice switch to going to port 3 once the cue is given there, then they are following the LED. If they continue to go to port 1, then they are following the landmarks in the arena. 
- **Phase 5:**
*Aim*: Introduction of port 3 consistently.
*Task*: Mice are given the cue at port 3, and are allowed to make mistakes without receiving an error signal.
*Performance evaluation*: This phase is only brief, and is moved on from when mice are showing that they can go to port 3 consistenly and aren't just checking randomly.
- **Phase 7:**
*Aim*: To teach mice to consistently follow the LED to get the reward.
*Task*: The cue and correct port are switched randomly between port 1 and port 3. Mice are given error signal for incorrect port touches.
*Performance evalutation*: Mice are moved on when they are consistenly getting the task correct and following the LED location.
- **Phase 9:**
*Aim*: To teach the mice that the reward is potentially given at all 6 ports.
*Task*: Cue randomly switched between ports 1 to 6.
- **Phase 4b:**
*Aim*: Teaching the mice audio trials.
*Task*: The cue is randomly switched between a visual cue at port 1 and an audio cue. Mice must go to port 1 to receive reward in both cases.
- **Phase 4c:**
*Aim*: Continued audio trial education.
*Task*: Mice are only given audio trials and must go to port 1 to receive reward. Error signals are given for incorrect port touches.
- **Phase 9b:**
*Aim*: To teach mice a full version of the mixed cue task.
*Task*: Visual cues are given at all ports, and audio cues are given on random trials. If the visual cue is given, micem ust go to the corresponding port, if an audio cue is given, mice must go to port 1. The proportion of audio:visual trials is set by the experimenter. I used either 1:1 or 2 parts audio 3 parts visual on sessions. 
- **Phase 9c:**
*Aim*: To teach mice the short cue task.
*Task*: Same are 9b but the cue time is set to only give the cue for a short time, <1sec. Visual and audio cue time are identical. Same thing about proportion as 9b, I used 1:1 audio:visual on the end day of behaviour. 

### Cue time experiment:
Mice in this cohort went through an experiment where the performance on the task was evaluated for different visual cue durations. When the cue time is short and the cue is behind the mouse, they are unable to see if and so cannot tell which port to go to. However, for my future optogenetics experiments, it is important for me to target the inhibition time to specific times in a trial, for example specifically during cue presentation or perhaps just after. I therefore performed an experiment to evaluate the radial performance of mice with different cue lengths, to find a good middle point where mice are performing movements towards cues they are seeing instantly, instead of hunting around for the cue, but without having cues so short that performance drops even for ports in front of mice. 
Cue times used: 1000ms, 500ms, 300ms, 100ms. (needs to be checked from the data as didn't note this down.)
